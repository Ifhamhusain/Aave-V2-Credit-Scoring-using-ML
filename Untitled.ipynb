```python

with open("user-wallet-transactions.json", "r") as f:
    data = json.load(f)

df = pd.json_normalize(data)
print(df.columns)
df.head()
```

```python
print(df.columns.tolist())
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])

grouped = df.groupby('userWallet')

tx_count = grouped.size().rename('tx_count')

action_count = grouped['action'].nunique().rename('unique_actions')

action_freq = grouped['action'].value_counts().unstack().fillna(0)
action_freq.columns = [f'action_{col}' for col in action_freq.columns]

activity_span = (grouped['timestamp'].max() - grouped['timestamp'].min()).dt.days.rename('active_days')

mean_interval = grouped['timestamp'].apply(lambda x: x.sort_values().diff().mean()).dt.total_seconds().fillna(0).rename('mean_time_diff_sec')

features = pd.concat([tx_count, action_count, action_freq, activity_span, mean_interval], axis=1).reset_index()
features = features.fillna(0)
features.head()
```

```python
scaler = MinMaxScaler()
scaled_values = scaler.fit_transform(features.drop(columns='userWallet'))
scaled_df = pd.DataFrame(scaled_values, columns=features.columns[1:])
scaled_df['userWallet'] = features['userWallet']

scaled_df['score'] = (
    0.2 * scaled_df['tx_count'] +
    0.2 * scaled_df['unique_actions'] +
    0.15 * scaled_df.get('action_repay', 0) +
    0.15 * scaled_df.get('action_deposit', 0) +
    0.1 * scaled_df.get('action_redeemunderlying', 0) +
    0.1 * scaled_df['active_days'] +
    0.1 * (1 - scaled_df.get('action_liquidationcall', 0))  # lower is better
)

scaled_df['credit_score'] = (scaled_df['score'] - scaled_df['score'].min()) / (scaled_df['score'].max() - scaled_df['score'].min()) * 1000
result = scaled_df[['userWallet', 'credit_score']].rename(columns={'userWallet': 'wallet'})
result.head(10000)
```

```python
result.to_csv('wallet_scores.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution Plot
plt.figure(figsize=(10, 5))
sns.histplot(result['credit_score'], bins=20, kde=True)
plt.title("Wallet Credit Score Distribution")
plt.xlabel("Score (0–1000)")
plt.ylabel("Wallet Count")
plt.grid(True)
plt.show()
```
